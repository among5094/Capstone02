```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np

# 이미지 불러오기 (예: 'sample_image.jpg' 라는 이미지 파일)
img_path = "C:/Users/among/OneDrive/바탕 화면/tumbler_img_for_AI/tumbler_2.img.jpg"
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(150, 150))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, 0)  # 배치를 만듭니다 (배치 사이즈 1)

# ImageDataGenerator 설정
datagen = ImageDataGenerator(
    rotation_range=40,        # 회전
    width_shift_range=0.2,    # 가로로 이동
    height_shift_range=0.2,   # 세로로 이동
    shear_range=0.2,          # 전단 변환
    zoom_range=0.2,           # 확대/축소
    horizontal_flip=True,     # 좌우 반전
    fill_mode='nearest'       # 빈 공간을 채우는 방식
)

# 이미지 증강 예시 보기
def plot_images(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20, 20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

# 증강된 이미지 생성
augmented_images = [datagen.flow(img_array, batch_size=1)[0].astype('uint8') for i in range(5)]

# 증강된 이미지 시각화
plot_images([img_array[0].astype('uint8')] + [img[0] for img in augmented_images])
```

## 최종 이미지 회전 코드

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array, load_img, save_img
import matplotlib.pyplot as plt
import numpy as np
import os

# 이미지 불러오기 (예: 'sample_image.jpg' 라는 이미지 파일)
img_path = "C:/Users/among/OneDrive/바탕 화면/tumbler_img_for_AI/tumbler_100.jpg" #webp  png  jpg
img = load_img(img_path, target_size=(150, 150))
img_array = img_to_array(img)

# 회전된 이미지 저장 경로 설정
save_dir = "C:/Users/among/OneDrive/바탕 화면/tumbler_img_for_AI/rotated_images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 회전 함수
def rotate_and_save(image_array, angle, save_dir, img_num):
    rotated_image = tf.image.rot90(image_array, k=img_num)  # k에 따라 90도, 180도, 270도 회전
    save_path = os.path.join(save_dir, f"tumbler_100_rotate_{angle}.jpg")
    save_img(save_path, rotated_image.numpy().astype('uint8'))

# 원본 이미지 시각화
plt.imshow(img_array.astype('uint8'))
plt.axis('off')
plt.show()

# 90도, 180도, 270도 회전 및 저장
rotate_and_save(img_array, 90, save_dir, 1)   # 90도 회전
rotate_and_save(img_array, 180, save_dir, 2)  # 180도 회전
rotate_and_save(img_array, 270, save_dir, 3)  # 270도 회전
```

## 가우시안 노이즈 추가 코드

# 1트

```python
import cv2
import numpy as np
import os  # 파일 경로 처리를 위한 os 모듈

# 가우시안 노이즈를 생성하여 이미지에 추가하는 함수
def make_noise(std, gray):
    # 이미지의 높이와 너비 추출
    height, width = gray.shape
    
    # 노이즈가 추가될 이미지와 같은 크기의 빈 배열 생성 (float 형식으로 생성)
    img_noise = np.zeros((height, width), dtype=np.float)
    
    # 각 픽셀에 대해 노이즈 추가
    for i in range(height):
        for a in range(width):
            # 가우시안 분포를 따르는 무작위 값 생성
            make_noise = np.random.normal()  # 평균 0, 표준편차 1의 가우시안 노이즈 생성
            
            # 가우시안 노이즈 값을 주어진 표준편차(std)로 조절
            set_noise = std * make_noise
            
            # 원본 이미지의 각 픽셀 값에 노이즈 값을 더함
            img_noise[i][a] = gray[i][a] + set_noise
    
    # 노이즈가 추가된 이미지 반환
    return img_noise


# 프로그램의 메인 실행 함수
def run():
    # 노이즈를 추가할 이미지 경로
    image_path = 'Augmentation/tumbler_img_for_AI/tumbler_1.jpg'
    
    # 이미지 파일 읽기 (지정한 경로에서 파일을 불러옴)
    img = cv2.imread(image_path)
    
    # 이미지가 제대로 로드되지 않은 경우 오류 메시지 출력 후 함수 종료
    if img is None:
        print(f"Error: 이미지 파일을 찾을 수 없습니다. 경로를 확인하세요: {image_path}")
        return

    # 이미지를 그레이스케일(흑백)로 변환
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 이미지의 높이와 너비 추출
    height, width = gray.shape

    # 노이즈의 표준편차 설정 (노이즈 강도를 조절하는 변수, 값이 클수록 노이즈가 강해짐)
    std = 15
    
    # 노이즈가 추가된 이미지 생성
    img_noise = make_noise(std, gray)

    # 결과 이미지를 저장할 파일 이름과 경로 설정
    output_path = os.path.join(os.path.dirname(image_path), 'tumbler_1_noise.jpg')

    # 노이즈가 추가된 이미지를 지정된 경로에 저장
    cv2.imwrite(output_path, img_noise.astype(np.uint8))
    
    print(f"노이즈가 추가된 이미지가 저장되었습니다: {output_path}")

    # 원본 그레이스케일 이미지와 노이즈가 추가된 이미지를 화면에 표시
    cv2.imshow("original", gray)  # 원본 이미지 출력
    cv2.imshow('noise', img_noise.astype(np.uint8))  # 노이즈가 추가된 이미지 출력
    
    # 키 입력 대기 (화면 표시 후 프로그램이 종료되지 않도록 대기)
    cv2.waitKey(0)


# 프로그램 실행
run()
```

# 2트

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array, load_img, save_img
import matplotlib.pyplot as plt
import numpy as np
import os

# 이미지 불러오기 (예: 'sample_image.jpg' 라는 이미지 파일)
img_path = "C:/Users/among/OneDrive/바탕 화면/tumbler_img_for_AI/tumbler_71.jpg"  # webp, png, jpg
img = load_img(img_path, target_size=(150, 150))
img_array = img_to_array(img)

# 회전된 이미지 및 노이즈 추가된 이미지 저장 경로 설정
save_dir = "C:/Users/among/OneDrive/바탕 화면/tumbler_img_for_AI/noise_and_rotated_images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 가우시안 노이즈 추가 함수
def add_gaussian_noise(image_array, std):
    noise = np.random.normal(0, std, image_array.shape)  # 가우시안 노이즈 생성
    noisy_image = image_array + noise  # 노이즈를 이미지에 더함
    noisy_image = np.clip(noisy_image, 0, 255)  # 값 범위를 0~255로 제한
    return noisy_image

# 회전 및 노이즈 추가 후 저장하는 함수
def rotate_add_noise_and_save(image_array, angle, std, save_dir, img_num):
    # 이미지를 회전
    rotated_image = tf.image.rot90(image_array, k=img_num)  # k에 따라 90도, 180도, 270도 회전
    
    # 회전된 이미지에 가우시안 노이즈 추가
    noisy_rotated_image = add_gaussian_noise(rotated_image.numpy(), std)
    
    # 이미지 저장 경로 설정
    save_path = os.path.join(save_dir, f"tumbler_100_rotate_{angle}_noise.jpg")
    
    # 노이즈가 추가된 이미지 저장
    save_img(save_path, noisy_rotated_image.astype('uint8'))

# 원본 이미지 시각화
plt.imshow(img_array.astype('uint8'))
plt.axis('off')
plt.show()

# 가우시안 노이즈 표준편차 (std) 설정
std = 25  # 노이즈 강도

# 90도, 180도, 270도 회전 및 노이즈 추가 후 저장
rotate_add_noise_and_save(img_array, 90, std, save_dir, 1)   # 90도 회전 + 노이즈
rotate_add_noise_and_save(img_array, 180, std, save_dir, 2)  # 180도 회전 + 노이즈
rotate_add_noise_and_save(img_array, 270, std, save_dir, 3)  # 270도 회전 + 노이즈
```
